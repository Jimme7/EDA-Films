# LIMPIEZA DE DATOS

```python
import pandas as pd
```

```python
films = pd.read_csv('../Data/films.csv', index_col= 0)
genre_films = pd.read_csv('../Data/films_genre.csv', index_col= 0)
series = pd.read_csv('../Data/series.csv', index_col= 0)
genre_series = pd.read_csv('../Data/series_genre.csv', index_col= 0)
```

```python
#Vemos si hay registros duplicados
len(films[films.duplicated()])
```

```python
#La columna con el id de cada película debería tener valores únicos, vamos a comprobarlo:
len(films['film_id'][films['film_id'].duplicated()])
```

```python
#Vamos a ver cuantos valores distintos tenemos en la columna original_language
films['original_language'].value_counts()
```

```python
#La columna género_id de cada DF es una lista, vamos a quedarnos con el primer género como si fuese el género principal

films ['genre_ids'] = films['genre_ids'].str.replace('[','').str.replace(']', '').str.split(',').str[0]
```

```python
films
```

```python
#Como podemos comprobar, al hacer esto hay filas que quedan con espacios vacíos
mask = films['genre_ids'] == ''
len(films[mask])
```

```python
#Como no son muchas filas, vamos a borrarlas
films.drop(films[mask].index, inplace= True)
films.reset_index(inplace=True, drop= True)
```

```python
films
```

```python
films['genre_ids'] = films['genre_ids'].astype(int)
```

```python
pd.merge(films,genre_films, left_on= 'genre_ids', right_on= 'id')
```
