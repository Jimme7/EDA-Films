# LIMPIEZA DE DATOS

```python
import pandas as pd
```

```python
films = pd.read_csv('../Data/films.csv', index_col= 0)
genre_films = pd.read_csv('../Data/films_genre.csv', index_col= 0)
series = pd.read_csv('../Data/series.csv', index_col= 0)
genre_series = pd.read_csv('../Data/series_genre.csv', index_col= 0)
```

----

### `LIMPIEZA DEL DF DE PELÍCULAS`

```python
films.head()
```

```python
films.info()
```

```python
genre_films.head()
```

```python
genre_films.info()
```

```python
#Vemos si hay registros duplicados
len(films[films.duplicated()])
```

```python
#Borramos los registros duplicados
films.drop_duplicates(inplace=True)
```

```python
#La columna con el id de cada película debería tener valores únicos, vamos a comprobarlo:
len(films['film_id'][films['film_id'].duplicated()])
```

```python
#Borramos valores nulos
films.dropna(inplace=True)
```

```python
#La columna género_id de cada DF es una lista, vamos a quedarnos con el primer género como si fuese el género principal

films ['genre_ids'] = films['genre_ids'].str.replace('[','').str.replace(']', '').str.split(',').str[0]
```

```python
films
```

```python
#Al hacer esto hay filas que quedan con espacios vacíos
mask = films['genre_ids'] == ''
len(films[mask])
```

```python
#Como no son muchas filas, vamos a borrarlas
films.drop(films[mask].index, inplace= True)
films.reset_index(inplace=True, drop= True)
```

```python
films
```

```python
#lo convertimos a dato numérico
films['genre_ids'] = films['genre_ids'].astype(int)
```

```python
#Y hacemos un merge con el df de géneros para poder unir cada género a su género_id
films = pd.merge(films,genre_films, left_on= 'genre_ids', right_on= 'id')
films.head(5)
```

```python
#Vamos a cambiar la columna release_date a tipo fecha
films['release_date'] = pd.to_datetime(films['release_date'])
```

```python
#Vamos a crear a partir de ella una columna con los meses, y otra con los años
films['release_year'] = films['release_date'].dt.year
films['release_month'] = films['release_date'].dt.month_name()
```

```python
films
```

```python
#Y a redondear la columna de popularidad a números enteros
films['popularity'] = round(films['popularity'])
films['popularity'] = films['popularity'].astype(int)
```

```python
#Por último, borramos la columna de los id de los géneros, pues solo la necesitábamos para poder unir estos
films.drop(columns=['id', 'genre_ids'])
```

```python
films.reset_index(drop=True, inplace=True)
```

```python
films.to_csv('../Data/clean_films.csv')
```

### `LIMPIEZA DEL DF DE SERIES`

```python
series.head()
```

```python
series.info()
```

```python
genre_series.head()
```

```python
genre_series.info()
```

```python
#Vamos a borrar registros con valores nulos y registros duplicados

series.drop_duplicates(inplace=True)
series.dropna(inplace=True)
```

```python
#La columna género_id de cada DF es una lista, vamos a quedarnos con el primer género como si fuese el género principal
series ['genre_id'] = series['genre_id'].str.replace('[','').str.replace(']', '').str.split(',').str[0]
```

```python
#Borramos las filas donde no hay registro de género
series = series[~(series['genre_id'] == '')]
```

```python
#unimos los géneros al df principal utilizando merge, primero pasando la columna de id a int
series['genre_id'] = series['genre_id'].astype(int)
series = pd.merge(series,genre_series, left_on= 'genre_id', right_on= 'id')
```

```python
#Nos quedamos con el primer país como país principal y quitamos el formato lista y eliminamos registros vacíos

series['origin_country'] = series['origin_country'].str.replace('[','').str.replace(']', '').str.replace("'", '').str.split(',').str[0]
series = series[~(series['origin_country'] == '')]
```

```python
#Redondeamos la popularidad a un número entero
series['popularity'] = round(series['popularity'])
series['popularity'] = series['popularity'].astype(int)
```

```python
#Convertimos la columna first_air_date a formato fecha y creamos la columna mes y año
series['first_air_date'] = pd.to_datetime(series['first_air_date'])
series['release_year'] = series['first_air_date'].dt.year
series['release_month'] = series['first_air_date'].dt.month_name()
```

```python
#Por último, borramos las columnas de id de género y reseteamos el índice
series.drop(columns=['genre_id', 'id'], inplace=True)
series.reset_index(drop=True, inplace=True)
```

```python
#Lo guardamos en otro DF
series.to_csv('../Data/clean_series.csv')
```
