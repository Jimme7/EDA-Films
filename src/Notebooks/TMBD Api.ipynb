```python
#Ahora vamos a crear un bucle while para obtener la información de las 100 primeras páginas
i = 1
serie_id = []
name = []
genre_id =[]
origin_country = []
original_language =[]
popularity = []
vote_average = []
vote_count = []

while i < 100:
  url = "https://api.themoviedb.org/3/discover/tv?include_adult=false&include_null_first_air_dates=false&language=en-US&page={}&sort_by=popularity.desc".format(i)
  headers = {
    "accept": "application/json",
    "Authorization": "Bearer eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiI1YTRjNWU0MGNlZWM2NTMwZjBhZmNkZjVmNjMwOGYyOSIsInN1YiI6IjYzZjYwYzY5OGMyMmMwMDBjZTNlNjMxMCIsInNjb3BlcyI6WyJhcGlfcmVhZCJdLCJ2ZXJzaW9uIjoxfQ.4k-HhqE--f_0tsWbbX2iaMeFAgyAOVSPynKbQ7rntJM"
    }
  response = requests.get(url, headers=headers)
  j = response.json()
  
  for z in range(20):
    serie_id.append(j['results'][z]['id'])
    name.append(j['results'][z]['name'])
    genre_id.append(j['results'][z]['genre_ids'])
    origin_country.append(j['results'][z]['origin_country'])
    popularity.append(j['results'][z]['popularity'])
    vote_average.append(j['results'][z]['vote_average'])
    vote_count.append(j['results'][z]['vote_count'])
  
  i += 1
```

```python
data = {
    'serie_id': serie_id,
    'name': name,
    'genre_id': genre_id,
    'origin_country': origin_country,
    'popularity': popularity,
    'vote_average': vote_average,
    'vote_count': vote_count
}

series = pd.DataFrame(data)
series
```

```python
# Y guardamos el df obtenido para posteriormente trabajar con el
series.to_csv("../Data/series.csv")
```

### `Extraemos los datos acerca de las películas:`

```python
i = 1
film_id = []
title = []
genre_id =[]
original_language =[]
release_date = []
popularity = []
vote_average = []
vote_count = []

while i < 100:

    url = "https://api.themoviedb.org/3/discover/movie?include_adult=false&include_video=false&language=en-US&page={}&sort_by=popularity.desc".format(i)
    headers = {
    "accept": "application/json",
    "Authorization": "Bearer eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiI1YTRjNWU0MGNlZWM2NTMwZjBhZmNkZjVmNjMwOGYyOSIsInN1YiI6IjYzZjYwYzY5OGMyMmMwMDBjZTNlNjMxMCIsInNjb3BlcyI6WyJhcGlfcmVhZCJdLCJ2ZXJzaW9uIjoxfQ.4k-HhqE--f_0tsWbbX2iaMeFAgyAOVSPynKbQ7rntJM"
    }

    response = requests.get(url, headers=headers)
    j = response.json()

    for z in range(20):
        film_id.append(j['results'][z]['id'])
        title.append(j['results'][z]['title'])
        original_language.append(j['results'][z]['original_language'])
        genre_id.append(j['results'][z]['genre_ids'])
        popularity.append(j['results'][z]['popularity'])
        vote_average.append(j['results'][z]['vote_average'])
        vote_count.append(j['results'][z]['vote_count'])
        release_date.append(j['results'][z]['release_date'])

    i += 1

```

```python
data = {
    'film_id': film_id,
    'title': title,
    'original_language': original_language,
    'genre_ids': genre_id,
    'release_date': release_date,
    'popularity': popularity,
    'vote_average': vote_average,
    'vote_count': vote_count
}

films = pd.DataFrame(data)
films
```

```python
#Guardamos nuestro df en un csv para posteriormente trabajar con él

films.to_csv('../Data/films.csv')
```
